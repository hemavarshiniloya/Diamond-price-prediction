```python
data1=data[:10000]
```

```python
data1['color']=data1['color'].map({'E':0,'I':1,'J':2,'H':3,'F':4,'G':5,'D':6})
```

```python
data1['clarity'].unique()
```

```python
data1
```

```python

X=data1.drop(['x','y','z','table','price'],axis=1)
```

```python
X
```

```python
y=data1['price']
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.4,shuffle=True)
```

```python
from sklearn.linear_model import LinearRegression 
```

```python
dia=LinearRegression()
```

```python
dia.fit(X_train,y_train)
```

```python
prediction=dia.predict(X_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,prediction)
```
